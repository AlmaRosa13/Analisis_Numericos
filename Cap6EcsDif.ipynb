# **Método de Euler**

Dada la ecuación diferencial $y'=x+2y$ resolver por el método de Euler.

```python
import numpy as np
import matplotlib.pyplot as plt

def funcion(x,y):
  ec=x+2*y
  return ec

def solucion(x,y):
  sol=0.25*np.exp(2*x)-0.5*x-0.25
  return sol

h=float(input('Tamaño de paso: '))
s=float(input('¿Hasta que valor? '))
n=int((s/h)+1)
x=np.zeros(n)
y=np.zeros(n)
ys=np.zeros(n)
#x[0]=1
#y[0]=0.5
print(x[0],y[0])
for i in np.arange(1,n):
  y[i]=y[i-1]+(funcion(x[i-1],y[i-1]))*h
  x[i]=x[i-1]+h
  ys[i]=solucion(x[i-1],y[i-1])
  print(x[i],y[i])

# Grafica
fig = plt.figure()
plt.plot(x,y,'.--',color='red',label = 'Euler')
plt.plot(x,ys,'-*',color='blue',label='Función ')
plt.grid()
plt.legend()
plt.title('Método de Euler con '+ str(n)+' pasos ')
plt.xlabel('tiempo')
plt.ylabel( '-')
plt.show()
```

Dada la ecuación diferencial $\frac{dy}{dx}=0.1 \sqrt{y}+0.4x^2$ resolver por el método de Euler.

```python
import numpy as np
import matplotlib.pyplot as plt

def funcion(x,y):
  ec=0.1*np.sqrt(y)+0.4*x**2
  return ec

h=float(input('Tamaño de paso: '))
s=float(input('¿Hasta que valor? '))
n=int((s/h)+1)
x=np.zeros(n)
y=np.zeros(n)

#x[0]=1
#y[0]=0.5
print(x[0],y[0])
for i in np.arange(1,n):
  y[i]=y[i-1]+(funcion(x[i-1],y[i-1]))*h
  x[i]=x[i-1]+h
  print(x[i],y[i])

# Grafica
fig = plt.figure()
plt.plot(x,y,'.--',color='red',label = 'Euler')
plt.grid()
plt.legend()
plt.title('Método de Euler con '+ str(n)+' pasos ')
plt.xlabel('tiempo')
plt.ylabel( '-')
plt.show()
```

Emplea el método de Euler para resolver la ecuación diferencial $\frac{dh}{dt}=-2.7684 \times 10^-3 \sqrt{h}$ resolver por el método de Euler.

```python
import numpy as np
import matplotlib.pyplot as plt

# Metodo de Euler
def euler(f,x0,y0,x1,n):
  h=(x1-x0)/n # tamano de paso
  xi=np.zeros(n+1) # vector de x variable independiente
  yi=np.zeros(n+1) # vector de y variable dependiente
  xi[0]= x0 # tiempo iniicial
  yi[0]= y0 # concentracion inicial

  for i in range (n):
    yi[i+1]= yi[i]+h*f(xi[i],yi[i])
    xi[i+1]= xi[i]+h
  return xi,yi # Vector de valores calculados

def f(x,y):
  return -2.7684*10**-3*np.sqrt(y)

def f2(x):
  return np.exp(-2*x)*1.5

def main():
  x0 =0 # valor inicial de tiempo
  y0 =4 # valor inicial de la concentracion
  x1 =1444 # valor final del tiempo
  n =50 # numero de pasos
  # llamada a la funcion euler
  x,y= euler(f,x0,y0,x1,n)
  print('x = ',x)
  print ('y = ',y)

  # Grafica
  fig = plt.figure()
  plt.plot(np.linspace(0,1600,51),y,'.--',label = 'Euler')
#  plt.plot(np.linspace(x0,x1,50),f2(np.linspace(x0,x1,50)),label='Función ')
  plt.grid()
  plt.legend()
  plt.title('Método de Euler con '+ str(n)+' pasos ')
  plt.xlabel('tiempo')
  plt.ylabel( 'Concentración')
  plt.show()

if __name__=="__main__":main()
```
